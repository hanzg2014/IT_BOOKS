```python
# USAGE
# python equalize.py --image ../images/beach.png

# Import the necessary packages
import numpy as np
import argparse
import cv2
from matplotlib import pyplot as plt
```

```python
# Load the image and convert it to grayscale
image = cv2.imread("../images/beach.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
hist1 = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(hist1, color = "blue")

# Apply histogram equalization to stretch the constrast
# of our image
eq = cv2.equalizeHist(image)
hist2 = cv2.calcHist([eq], [0], None, [256], [0, 256])
plt.plot(hist2, color = "red")
```

```python
# Show our images -- notice how the constrast of the second
# image has been stretched
cv2.imshow("Histogram Equalization", np.hstack([image, eq]))
cv2.waitKey(0)
```
